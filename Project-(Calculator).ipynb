```python
while True:
    print("welcome to the calculator")
    print("1 : addition")
    print("2 : substraction")
    print("3 : division")
    print("4 : multiplication")
    print("5 : quit")
    c=input("enter your choice:")

    if(c=="1"):
        x=int(input("enter the number :"))
        y=int(input("enter the number :"))
        print(f"{x}+{y}={x+y}")
    elif(c=="2"):
        x=int(input("enter the number :"))
        y=int(input("enter the number :"))
        print(f"{x}-{y}={x-y}")
    elif(c=="3"):
        x=int(input("enter the number :"))
        y=int(input("enter the number :"))
        print(f"{x}%{y}={x%y}")
    elif(c=="4"):
        x=int(input("enter the number :"))
        y=int(input("enter the number :"))
        print(f"{x}*{y}={x*y}")
    elif(c=="5"):
        d=input("do you really wan't exit(y/n):")
        if(d.lower()=="y"):
            break
        else:
            continue
    else:
        print("invalid operator")
```
